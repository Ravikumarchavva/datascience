```python
matches=pl.read_csv(r"D:\Ravikumar\Git\ravikumar\statistics\testing\ipl\matches.csv")
matches
```

```python
matches=matches.drop("season","player_of_match","umpire1","umpire2",'umpire3',"city","win_by_runs","win_by_wickets").filter(pl.col("dl_applied")==0)
matches
```

```python
matches=matches.filter(pl.col("result")=="normal").drop("result","dl_applied","date")
matches
```

```python
deliveries=pl.read_csv(r"D:\Ravikumar\Git\ravikumar\statistics\testing\ipl\deliveries.csv")
deliveries=deliveries.filter(pl.col("match_id").is_in(matches["match_id"]))
deliveries
```

```python
deliveries=deliveries.drop("ball","batsman","non_striker",	"bowler","is_super_over","dismissal_kind","fielder","wide_runs"	,"bye_runs","legbye_runs","noball_runs","penalty_runs","batsman_runs","extra_runs")
deliveries
```

```python
deliveries=deliveries.with_columns(pl.col("player_dismissed").apply(lambda value : 0 if value is None else 1))
deliveries.fill_null(0)
```

```python
group_data=deliveries.groupby(pl.col("match_id"),pl.col("inning"),pl.col("over")).agg((pl.col("total_runs").sum()),(pl.col("player_dismissed").sum())).fill_null(0)#.select(deliveries["match_id"],deliveries["inning"],deliveries["batting_team"],deliveries["bowling_team"],deliveries["over"],deliveries["total_runs"],deliveries["player_dismissed"])
```

```python
group_data.filter((pl.col("match_id")==1) & (pl.col("inning")==2)).sort(pl.col("over"))
```

```python
powerplay_score_data=group_data.filter(pl.col("over")<=6).groupby(["match_id", "inning"]).agg(pl.col("total_runs").sum().alias("powerplay_runs"),pl.col("player_dismissed").sum().alias("powerplay_wickets")).sort("match_id")
powerplay_score_data
```

```python
final_score_data=group_data.groupby(["match_id", "inning"]).agg(pl.col("total_runs").sum().alias("final_runs"),pl.col("player_dismissed").sum().alias("final_wickets")).sort("match_id")
final_score_data
```

```python
score_data=powerplay_score_data.join(final_score_data,on=["match_id", "inning"],how="inner")
score_data
```

```python
deliveries=deliveries.drop("over","total_runs","player_dismissed")
deliveries=deliveries.with_columns(deliveries["match_id"],deliveries["inning"],deliveries["batting_team"],deliveries["bowling_team"]).groupby("match_id", "inning","batting_team","bowling_team").agg().sort("match_id", "inning")
deliveries
```
