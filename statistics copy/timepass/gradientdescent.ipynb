```python
import numpy as np
import pandas as pd
```

```python
dictionary={"x":[1,3,5],"y":[2,5,6]}
df=pd.DataFrame(dictionary,index=['1','2','3'])
df
```

```python
def gradient_descent(df):
    w=1
    b=0
    lr=0.01
    iteration=1000
    for i in range(iteration):
        y_hat=w*df.x+b
        dj_dw=sum(df.y-y_hat)*w/len(df.index)
        dj_db=sum(df.y-y_hat)/len(df.index)
        w=w-lr*dj_dw
        b=b-lr*dj_db
        print(f"y_hat:{y_hat} dj_dw:{dj_dw} dj_db:{dj_db} w:{w} b:{b}")
```

```python
gradient_descent(df)
```
